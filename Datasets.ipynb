```python
import pandas as pd
```

```python
store=pd.read_csv("/content/stores_data_set.csv")
feature=pd.read_csv("/content/Features_data_set.csv")
sale=pd.read_csv("/content/sales_data_set.csv")
```

```python
store_sales=pd.merge(sale,store,on="Store")
```

```python
feature.drop(columns="IsHoliday",inplace=True)
```

```python
final=pd.merge(store_sales,feature,on=["Store","Date"])
```

```python
final
```

```python
final.columns
```

```python
final.info()
```

```python
final["Date"]=pd.to_datetime(final["Date"],format="%d/%m/%Y")
```

```python
final["Year"]= final["Date"].dt.year
final["Month"]= final["Date"].dt.month
final["Day"]= final["Date"].dt.day
final["Day_of_week"] = final["Date"].dt.dayofweek
final["Week"]= final["Date"].dt.isocalendar().week
```

```python
final
```

```python
final.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
```

```python
final["IshHoliday"]=encoder.fit_transform(final[["IsHoliday"]])
final["Type"]=encoder.fit_transform(final[["Type"]])
```

```python
final.isnull().sum()
```

```python
final.fillna(0,inplace=True)
```

```python
final.duplicated().sum()
```

```python
final=final.drop("Date",axis=1)
```
