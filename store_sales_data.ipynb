```python
import pandas as pd
```

```python
sale=pd.read_csv("/content/sales_data_set.csv")
store=pd.read_csv("/content/stores_data_set.csv")
feature=pd.read_csv("/content/Features_data_set.csv")
```

```python
store.value_counts().sum(),
```

```python
sale.value_counts().sum(),
```

```python
feature.value_counts().sum()
```

```python
store_sale=pd.merge(sale,store,on="Store")
```

```python
store_sale.columns,feature.columns
```

```python
store_sale_feature=pd.merge(store_sale,feature,on=["Store","Date","IsHoliday"])
```

```python
store_sale_feature.columns
```

```python
ssf=store_sale_feature.copy()
```

```python
ssf
```

```python
ssf.isnull().sum()
```

```python
ssf.info()
```

```python
ssf.fillna(0,inplace=True)
```

```python
ssf.info()
```

```python
final=ssf.copy()
```

```python
final
```

```python
final["Date"]=pd.to_datetime(final["Date"],format="%d/%m/%Y")
```

```python
final["Year"]= final["Date"].dt.year
final["Month"]= final["Date"].dt.month
final["Day"]= final["Date"].dt.day
final["Day_of_week"] = final["Date"].dt.dayofweek
final["Week"]= final["Date"].dt.isocalendar().week
```
